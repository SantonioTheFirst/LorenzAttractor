```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from PIL import Image
```

```python
tmin = 0
tmax = 100
dt = 0.00001
sigma = 10.
rho   = 28.
beta  = 8./3.


def lorenz(xpoints, ypoints, zpoints):
    for i in range(1, xpoints.shape[0]):
        xpoints[i] = xpoints[i - 1] + (sigma * (ypoints[i - 1] - xpoints[i - 1])) * dt
        ypoints[i] = ypoints[i - 1] + (xpoints[i - 1] * (rho - zpoints[i - 1]) - ypoints[i - 1]) * dt
        zpoints[i] = zpoints[i - 1] + (xpoints[i - 1] * ypoints[i - 1] - beta * zpoints[i - 1]) * dt
    return xpoints, ypoints, zpoints
```

```python
values = [(0.5, 0., 0.), (0.5, 0.5, 0.), (0.5, 0.5, 0.5)] #, (0.7, 0.8, 0.9), (1., 1., 1.)]
fig = plt.figure(figsize=(12, 9))
for value in values:
    x, y, z = value

    xpoints = np.zeros(int((tmax - tmin) / dt), dtype=float)
    ypoints = np.zeros_like(xpoints)
    zpoints = np.zeros_like(xpoints)
    xpoints[0] = x
    ypoints[0] = y
    zpoints[0] = z

    xpoints, ypoints, zpoints = lorenz(xpoints, ypoints, zpoints)

    ax = fig.gca(projection='3d')
    ax.xaxis.set_pane_color((1,1,1,1))
    ax.yaxis.set_pane_color((1,1,1,1))
    ax.zaxis.set_pane_color((1,1,1,1))
    ax.plot(xpoints, ypoints, zpoints, alpha=0.7, linewidth=1.0)
plt.tight_layout()
plt.title('Lorenz Attractor', {'fontsize': 20})
plt.savefig('LorenzAttractor.png', dpi=200)
plt.show()
```
